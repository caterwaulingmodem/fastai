```python
from os import listdir
from os.path import isfile, join
from sklearn.cross_validation import train_test_split
import random

all_files = [data_file for data_file in listdir('train') if isfile(join('train', data_file))]

random.shuffle(all_files)
x_train, x_valid = train_test_split(all_files, test_size=0.10)
print(len(x_train))
print(len(x_valid))
```

```python
import pandas as pd
labels_filename = 'labels.csv'
labels = pd.read_csv(labels_filename, header=0, index_col=None, encoding="utf-8")
labels.head()
```

```python
unique_breeds = labels.breed.unique()
len(unique_breeds)
```

```python
def create_label_dirs(path):
    for breed in unique_breeds:
        try:
            os.mkdir(path+'/'+breed)
        except OSError as e:
            print(e)
```

```python
for folder in ['train', 'valid' , 'sample/train', 'sample/valid']:
    create_label_dirs(folder)
```
